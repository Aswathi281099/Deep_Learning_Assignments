**Recurrent Neural Network (RNN)**

Objective:

The objective of a Recurrent Neural Network (RNN) is to process sequential data by remembering previous information and using it to make accurate predictions or classifications.

Step 1: Load the Dataset

```python
!pip install datasets

# Import required libraries
from datasets import load_dataset
import pandas as pd

# Load AG News dataset
dataset = load_dataset("wangrongsheng/ag_news")

# Convert into DataFrame
train_df = pd.DataFrame(dataset["train"])
test_df = pd.DataFrame(dataset["test"])

# Display first few rows
print(train_df.head())
```

Step 2: Clean and Normalize the Text

```python
import re

train_df["text"] = train_df["text"].str.lower()
train_df["text"] = train_df["text"].apply(lambda x: re.sub(r'[^a-zA-Z ]', '', x))
```

Step 3: Tokenize the Strings into Numbers

```python
from tensorflow.keras.preprocessing.text import Tokenizer

tokenizer = Tokenizer(num_words=10000)
tokenizer.fit_on_texts(train_df["text"])

X_train = tokenizer.texts_to_sequences(train_df["text"])
X_test = tokenizer.texts_to_sequences(test_df["text"])
```

Step 4: Pad and Truncate Sequences

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

X_train = pad_sequences(X_train, maxlen=50)
X_test = pad_sequences(X_test, maxlen=50)
```

Step 5: Convert Labels to Categorical Vectors

```python
from tensorflow.keras.utils import to_categorical
y_train = to_categorical(train_df["label"])
y_test = to_categorical(test_df["label"])
```

Step 6: Define the RNN Model Architecture


```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense

model = Sequential()
model.add(Embedding(input_dim=10000, output_dim=64, input_length=50))
model.add(SimpleRNN(64))
model.add(Dense(4, activation="softmax"))
```

Step 7: Compile and Train the Model

```python
model.compile(optimizer="adam",
              loss="categorical_crossentropy",
              metrics=["accuracy"])

model.fit(X_train, y_train,
          epochs=5,
          validation_split=0.2)
```

Step 8: Read and Evaluate the Target Accuracy

```python
loss, accuracy = model.evaluate(X_test, y_test)

print("Accuracy:", accuracy)
```
